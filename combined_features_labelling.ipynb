```python
import pandas as pd

labels = pd.read_csv("bubble_labels.csv")
int_data = pd.read_csv("integrated_data.csv")

labels_filtered = labels[["timestamp", "symbol", "gsadf_label_95", "gsadf_label_90"]].rename(columns = {"timestamp": "date"})

merged = int_data.merge(labels_filtered, on = ["date", "symbol"], how = "left")

merged.to_csv("integrated_data_with_labels.csv", index = False)
```
