
# Data Standardisation

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
# 1. Load dataset
df = pd.read_csv('data/df_features_with_labels.csv')
```

```python
# 2. One-hot encoding for quarter (year column absent — skipped)
df = pd.get_dummies(df, columns=['quarter'], prefix='quarter', drop_first=True, dtype=int)
if 'year' in df.columns:
    df = pd.get_dummies(df, columns=['year'], prefix='year', drop_first=True, dtype=int)
```

```python
# 4. Drop redundant columns
columns_to_exclude = ['gsadf_label_95', 'gsadf_label_90', 'fear_greed_classification']
df = df.drop(columns=columns_to_exclude)
```

```python
pd.set_option('display.max_columns', None)
df
```

```python
scaler = StandardScaler()
df_scaled = df.copy()

date_cols = ['day_of_week', 'month']
binary_cols = ['is_month_start', 'is_month_end', 'risk_off_signal', 'vix_regime', 'extreme_greed', 'greed', 'extreme_fear', 'fear', 'quarter_2', 'quarter_3', 'quarter_4', 'year_2019', 'year_2020', 'year_2021', 'year_2022', 'year_2023', 'year_2024', 'year_2025']
output_cols = ['label', 'label_lead']
not_feature_cols = date_cols + binary_cols + output_cols + ['symbol', 'date']
feature_cols = [c for c in df.columns if c not in not_feature_cols]

# remove +inf or -inf values to avoid standardiser exploding and use ffill or bfill to replace the value
df.replace([np.inf, -np.inf], np.nan, inplace=True)

df[feature_cols] = df.groupby('symbol')[feature_cols].ffill()
df[feature_cols] = df.groupby('symbol')[feature_cols].bfill()

df[feature_cols] = df[feature_cols].astype(float) # convert to float64 for fear & greed streak (int64)
df_scaled[feature_cols] = df_scaled[feature_cols].astype(float) # ensure df_scaled can hold float values

for coin in df['symbol'].unique():
    mask = df['symbol'] == coin
    df_scaled.loc[mask, feature_cols] = scaler.fit_transform(df.loc[mask, feature_cols])
```

```python
# 7. Use cyclical standardisation for date-related columns, drop year
df_scaled['month_sin'] = np.sin(2 * np.pi * df_scaled['month'] / 12)
df_scaled['month_cos'] = np.cos(2 * np.pi * df_scaled['month'] / 12)

df_scaled['day_of_week_sin'] = np.sin(2 * np.pi * df_scaled['day_of_week'] / 7)
df_scaled['day_of_week_cos'] = np.cos(2 * np.pi * df_scaled['day_of_week'] / 7)
```

```python
# 8. Drop raw date_cols
df_scaled = df_scaled.drop(columns=date_cols)
```

```python
# 9. Export to csv
df_scaled.to_csv('data/df_features_with_label_standardised.csv', index=False)
```
