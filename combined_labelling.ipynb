```python
import pandas as pd

labels = pd.read_csv("bubble_labels.csv")
features = pd.read_csv("df_features.csv")

labels_filtered = labels[["timestamp", "symbol", "gsadf_label_95", "gsadf_label_90"]].rename(columns = {"timestamp": "date"})

merged = features.merge(labels_filtered, on = ["date", "symbol"], how = "left")

merged.to_csv("df_features_with_labels.csv", index = False)
```
