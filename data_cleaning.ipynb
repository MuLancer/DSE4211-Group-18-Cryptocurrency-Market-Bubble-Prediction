```python
import pandas as pd
from functools import reduce
 
files = [
    'data/crypto_market_data_daily.csv',
    'data/google_trends_coin.csv',
    'data/google_trends_market.csv',
    'data/macro_data.csv',
    'data/sentiment_data.csv'
]
 
dfs = []
for f in files:
    df = pd.read_csv(f)
    df = df.rename(columns={'timestamp': 'date'}).dropna(subset=['date'])
    dfs.append(df)

merged = reduce(lambda l, r: pd.merge(l, r, on='date', how='outer'), dfs)
merged = merged.sort_values('date').reset_index(drop=True)
merged = merged[(merged['date'] >= '2018-05-01') & (merged['date'] <= '2025-01-01')]
print(f"Rows: {len(merged)} | Cols: {len(merged.columns)}")

# Check NA
merged.isna().sum()
```

```python
merged.to_csv('data/integrated_data.csv', index=False)
```
